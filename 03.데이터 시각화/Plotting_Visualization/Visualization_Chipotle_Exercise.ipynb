### Step 3. chipo 라는 변수에 DataFrame 을 할당하세요

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url, sep = "\t")
```

### Step 5. 구매 상위 5개 품목의 히스토그램을 생성하세요
- collections 의 Counter 를 사용하세요

```python
from collections import Counter
```

```python
x = chipo.item_name
```

```python
# chipo.groupby("item_name").count()
```

```python
df = pd.DataFrame()
```

```python
best_items = dict(Counter(chipo.item_name).most_common(5))
plt.bar(list(best_items.keys()),best_items.values())
plt.xticks(list(best_items.keys()), rotation= 30)
plt.xlabel("Items")
plt.ylabel("Number_of")
plt.title("Most ordered")
plt.show()
```

### Step 6. 주문가격당 주문횟수를 산점도를 생성하여 확인하시오
#### Hint: 주문 가격은 X축, 주문 횟수는Y축 

```python
chipo.item_price = [float(i[1:-1]) for i in chipo.item_price]
```

```python
# chipo.groupby("order_id").count()
```

```python
order = chipo.groupby("order_id").sum()
```

```python
plt.scatter(x = order.item_price, y = order.quantity)
plt.xlabel("order price")
plt.ylabel("ordered items")
plt.show()
```
