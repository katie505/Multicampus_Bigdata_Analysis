## 히스토그램

- 막대 그래프의 한 종류
- 양적 데이터를 범주로 사용할 때 유용하다.
    - 예를 들어, 무게, 길이, 시간, 점수 등 
    
- 숫자의 범위에 따라 나뉘어진 데이터를 막대 그래프로 보여줄 때 사용한다.
    - 양적 데이터들의 연속성을 반영하기 위해 막대 그래프들 사이의 간극이 없도록 표현한다.
    - 정해진 구간에 해당되는 항목의 개수나 빈도수를 보여주어 값의 분포를 쉽게 관찰
    
__참고__
- 질적 자료
    - 사칙 연산이 불가능
    - 비계측 자료(Nonmetric Data) 혹은 범주형 자료(Categorical Data)
    - 예) 직업, 종교, 성별

- 양적 자료
    - 사칙 연산 가능
    - 계측 자료(Metric Data) 혹은 숫자형 자료(Numerical Data)
    - 예) 몸무게, 용돈, 가격

````python
friends = [ 70,  65,  72,  63,  71,  64,  60,  64,  67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels =  ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
````

````python
plt.scatter(friends, minutes, s=30)
plt.annotate("a", 
             xy=(70,175))
````

````python
plt.scatter(friends, minutes)
for i, j, k in zip(labels, friends, minutes):
  plt.annotate(i, 
              xy=(j,k),
              xytext=(5, -5),
              textcoords="offset points")
plt.xlabel("friends")
plt.ylabel("minutes on the site")
plt.title("minutes per friends")
plt.show()
````

````python
x = np.random.rand(30)
y = np.random.rand(30)
area = (y*100)
colors = (y*10)

plt.scatter(x, y, s=area, c=colors)
````

````python
x = np.random.rand(30)
y = np.random.rand(30)
area = (y*100)
colors = (y*10)

plt.scatter(x, y, s=area, c=colors, alpha= 0.7 ,cmap="Spectral")
plt.colorbar()
# cmap : colormap을 의미 한다. 
# alpha : 투명도
````

#### 축 지정

* x 축: 1차 테스트 점수
* y 축: 2차 테스트 점수

#### 양 축의 척도(scale)가 다른 경우

````python
test_1_grades = [ 99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]
````

````python
plt.scatter(test_1_grades, test_2_grades)
plt.show()
````

결론: 1차 테스트, 2차 테스트 점수가 별 변동이 없는 것으로 보인다.

#### 양 축의 척도가 동일한 경우

````python
test_1_grades = [ 99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]
````

````python
plt.scatter(test_1_grades, test_2_grades)
plt.axis([50, 110, 50, 110])
plt.show
````

````python
plt.scatter(test_1_grades, test_2_grades)
plt.axis("equal")
plt.show
````

결론: 1차 테스트 점수가 높은 친구가 2차 테스트에서 보다 높은 점수를 얻었음을 알 수 있다.
