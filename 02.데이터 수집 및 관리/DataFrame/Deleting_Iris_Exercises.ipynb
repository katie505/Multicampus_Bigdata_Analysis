### Step 4. 다음과 같은 순서로 columns을 만드세요

```python
# 1. sepal_length (in cm)
# 2. sepal_width (in cm)
# 3. petal_length (in cm)
# 4. petal_width (in cm)
# 5. class
iris.columns = ['sepal_length', 'sepal_width', 'petal_length',
                'petal_width', 'class']
iris
```

### Step 5. 데이터 프레임의 각컬럼 마다 누락된 값이 존재 합니까?
