```python
import pandas as pd
import numpy as np
```

### Step 2. 다음은 dictionary로 주어진 dataset 입니다.

```python
# Create an example dataframe about a fictional army
raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'],
            'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'],
            'deaths': [523, 52, 25, 616, 43, 234, 523, 62, 62, 73, 37, 35],
            'battles': [5, 42, 2, 2, 4, 7, 8, 3, 4, 7, 8, 9],
            'size': [1045, 957, 1099, 1400, 1592, 1006, 987, 849, 973, 1005, 1099, 1523],
            'veterans': [1, 5, 62, 26, 73, 37, 949, 48, 48, 435, 63, 345],
            'readiness': [1, 2, 3, 3, 2, 1, 2, 3, 2, 1, 2, 3],
            'armored': [1, 0, 1, 1, 0, 1, 0, 1, 0, 0, 1, 1],
            'deserters': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
            'origin': ['Arizona', 'California', 'Texas', 'Florida', 'Maine', 'Iowa', 'Alaska', 'Washington', 'Oregon', 'Wyoming', 'Louisana', 'Georgia']}
```

### Step 3. army 라는 변수에 DataFrame을 할당하세요 

#### ('regiment', 'company', 'deaths'...) 과 같은 순서로 DataFrame이 생성되어야 합니다. 

```python
army = pd.DataFrame(raw_data)
```

### Step 4.  dataframe의 index로 'origin' column을 할당하세요

```python
army.set_index('origin', inplace = True)
army
```

### Step 5. veterans column을 출력하세요

```python
army['veterans']
```

### Step 6. 'veterans' 과 'deaths' columns 을 출력하세요

```python
army[['veterans', 'deaths']]
```

### Step 7. 모든 컬럼의 이름을 출력하세요

```python
army.columns
```

### Step 8. Maine 과 Alaska에서 'deaths', 'size', deserters' columns을 선택하여 출력하세요

```python
army.loc[['Maine', 'Alaska'], :][['deaths', 'size', 'deserters']]
```

### Step 9. 3 ~ 7행과 3 ~ 6열을 선택하세요

```python
army.iloc[2:7, 2:6]
```

### Step 10. 모든 columns에 대해 4행 이후의 모든 행을 출력하세요

```python
army.iloc[3:,:]
```

### Step 11. 모든 columns에 대해 4행까지만 출력하세요

```python
army.iloc[3:,:]
```

### Step 12. 3열부터 7열까지 선택하세요

```python
army.iloc[:, 2:7]
```

### Step 13. df.deaths가 50보다 큰 행을 선택하세요

```python
army[army.deaths > 50]
```

### Step 14. df.deaths가 500보다 크거나 50보다 작은 행을 선택하세요

```python
army[(army.deaths < 50) | (army.deaths > 500)]
```

### Step 15. regiments 가 "Dragoons"가 아닌 row를 출력하세요

```python
army[army.regiment != 'Dragoons']
```

### Step 16.  Texas 와 Arizona rows를 선택하세요

```python
army.loc[['Texas', 'Arizona'], :]
```

### Step 17. Arizona라는 행에서 세 번째 셀을 선택하세요
