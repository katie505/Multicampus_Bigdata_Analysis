### Step 1. 필요한 라이브러리 들을 import 하세요

```python
import pandas as pd
import numpy as np
```

### Step 2. 다음 주소의 파일을 import 하세요 [address](https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data). 

### Step 3. wine 라는 변수에 DataFrame 을 할당하세요

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'
wine = pd.read_csv(url)
```

### Step 4. 첫번째, 네번째, 여섯번째, 아홉번째, 열한번째, 열세번째, 열네번째 column 을 삭제하세요

```python
wine = wine.drop(wine.columns[[0, 3, 5, 8, 9, 12, 13]], axis = 1)
```

### Step 5.다음과 같은 순서로 columns을 만드세요

1) alcohol  
2) malic_acid  
3) alcalinity_of_ash  
4) magnesium  
5) flavanoids  
6) proanthocyanins  
7) hue 

```python
wine.columns = ['alcohol', 'malic_acid', 'alcalinity_of_ash',
                'magnesium', 'flavanoids', 'proanthocyanins', 'hue']

wine
```

### Step 6. alcohol의 처음 3개 행의 값을 NaN으로 설정하세요

```python
wine.alcohol[0:3] = np.nan
```

### Step 7. magnesium 의 3행과 4행의 값을 NaN으로 설정하세요

```python
wine.magnesium[3:5] = np.nan
```

### Step 8. NaN 값을 알코올의 경우 10, 마그네슘의 경우 100으로 채우세요

```python
wine.loc[wine['alcohol'] != wine['alcohol'], 'alcohol'] = 10
wine.loc[wine['magnesium'] != wine['magnesium'], 'magnesium'] = 100
wine
```

### Step 9. 데이터 프레임의 각컬럼 마다 누락된 값이 존재 합니까?

```python
wine.isnull().sum()
```

### Step 10. 0부터 9까지 10개의 난수로 구성된 np.array를 만드세요

```python
arr_1 = np.random.randint(0, 10, size = 10)
arr_1
```

### Step 11. alcohol column에 생성한 난수를 인덱스로 사용하여 각 셀에 NaN 값을 할당하세요

```python
wine.alcohol[arr_1] = np.nan
wine
```

### Step 12. DataFrame에 누락된 값이 몇 개입니까?

```python
wine.isnull().sum()
```

### Step 13. NaN값이 있는 행을 삭제 하세요

```python
wine2 = wine.dropna()
wine2
```

### Step 14. alcohol column에서 non-null 인 값만 출력하세요

```python
wine2.alcohol.notnull()
```

### Step 15. index 를 0 부터 시작하도록 reset하세요

```python
wine2.reset_index(drop=True)
```
